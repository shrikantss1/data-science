# Normality Test

The goal of this test is to identify if the given column is NORMALLY DISTRIBUTED

```python
import seaborn as sns
sns.kdeplot(data['R&D Spend'])
```

Goal: Check if R&DSpend is Normally distributed

```python
#Perform Hypothesis Test
#
# 1. Create Viable Question (Binary outcome)
#
#  Question: Test whether R&DSpend is Normally DIstributed
#
# 2. Convert the question into Hypothesis (H0 and H1)
#
#   H0 ------- R&DSpend is NOT NORMALLY DISTRIBUTED
#   Ha ------- R&DSpend is NORMALLY DISTRIBUTED
#
# 3. Select the statistical tool/formula to validate the hypothesis
#
#    Shapiro Test
#
# 4. Select/Determine the SL of the project | Alpha value = 0.05
#
# SL =0.05
#
# 5. Calc the p-value from the test and compare the same with SL to check who WON!
#







SL = 0.05
from scipy.stats import shapiro
_, pvalue = shapiro(data['R&D Spend'])

if pvalue >= SL:
  print("Alternate Pass - Ha - R&D Spend is NORMALLY DISTRIBUTED")
else:
  print("Null Pass - H0 --------------- R&D is NOT NORMALLY DISTRIBUTED")
```

# Test For Feature Elimination

```python
SL = 0.05
from scipy.stats import ttest_ind

_,pvalue = ttest_ind(data['R&D Spend'], data['Administration'])
if pvalue<=SL:
  print("Ha ------- R&D Spend and Admin Spend are statistically different. So PRESERVE both of them")
else:
  print("H0 ------- R&D Spend and Admin Spend are statistically same. So ELIMINATE one of them")
```

```python
SL = 0.05
from scipy.stats import shapiro

_,pvalue = shapiro(data['Price'])

if pvalue >= SL:
  print("Alternate Pass - Ha ------- Price is NORMALLY DISTRIBUTED")
else:
  print("Null Pass - H0 ------- Price is NOT NORMALLY DISTRIBUTED")

_,pvalue = shapiro(data['Car'])

if pvalue >= SL:
  print("Alternate Pass - Ha ------- Car is NORMALLY DISTRIBUTED")
else:
  print("Null Pass - H0 ------- Car is NOT NORMALLY DISTRIBUTED")
```

```python
#Goal: Check if R&D spend and Admin are statistically same

# Perform Hypothesis Testing
#
# 1. Create a viable question
#
#    Check if R&D spend and Admin are statistically same
#
# 2. Convert this question into Hypothesis
#
#  H0 ------- R&D Spend and Admin Spend are statistically same. So ELIMINATE one of them
#  Ha ------- R&D Spend and Admin Spend are statistically different. So PRESERVE both of them
#
# 3. Select the statistical test (Parametric Test)
#
#     wilcoxon
# 4. Select / Determin SL
#
#   SL = 0.05
#
# 5. Calc p value

SL = 0.05

from scipy.stats import wilcoxon

_,pvalue = wilcoxon(data['Price'], data['Car'])

if pvalue<=SL:
  print("Ha ------- Price and Car are statistically different. So PRESERVE both of them")
else:
  print("H0 ------- Price and Car are statistically same. So ELIMINATE one of them")
```

```python
ct = pd.crosstab(data['State'], data['Profit'])
```

```python
ct
```

```python
SL = 0.05

from scipy.stats import chi2_contingency

_,pvalue,_,_ = chi2_contingency(ct)

if pvalue <= SL:
  print("Ha --- State and Profit has Statistical relationship")
else:
  print("H0 --- State and Profit has NO Statistical relationship")
```
